# Demostration of "calling the API endpoint " 

### importing necessary libraries

```python

import json
import cv2
import requests
import numpy
from matplotlib import pyplot
```

## Conversion function 

```python
def str_to_numpy(face):
    """ function to convert the string to np array"""
    
    face_arr = ""
    
    for i in range(len(face)):
        if face[i] != '[' and face[i] != ']' and face[i] != ' ':
             face_arr+=face[i]

    face = numpy.asarray([int(i) for i in face_arr.split(',')])
    face = face.reshape(48,48)
    
    
    return face 
```

## Display function

```python
def present_output(response,img):
    """ printing the output """

    face = str_to_numpy(response['face'])
    mood = response['mood']
    songs = response['songs']
    
    print("_______orignal image___________")
    print('\n\n')
    pyplot.imshow(img)
    pyplot.show()


    print('\n\n')
    print("_______detected face ___________")
    print('\n\n')
    pyplot.imshow(face)
    pyplot.show()

    #
    print('\n\n')
    print("_______predictated mood  ___________")
    print('\n\n')
    print(mood)

    #predicted songs
    print('\n\n')
    print("_______predicted songs  ___________")
    print('\n\n')
    for s in songs:
        print(s)
        print('\n')
```

## Caller function 

please past your URL generated by the API gateway 


```python

# read the image
img = cv2.imread("Images/face11.jpg")

# convert it in requried size 
img = cv2.resize(img, (100,100), interpolation = cv2.INTER_AREA)


# calling the API endpoint 
data = {
        "data": img.tolist() 
       }

Endpoint_URL = "https://zcw43naxp5.execute-api.us-east-1.amazonaws.com/Prod"

response = requests.post( Endpoint_URL , json = data).json()


# printing the output 
present_output(response,img)

```
